### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda['Precio'] = tienda['Precio'].astype(int)
tienda2['Precio'] = tienda2['Precio'].astype(int)
tienda3['Precio'] = tienda3['Precio'].astype(int)
tienda4['Precio'] = tienda4['Precio'].astype(int)
```

```python
VentaT1 = tienda['Precio'].sum()
VentaT2 = tienda2['Precio'].sum()
VentaT3 = tienda3['Precio'].sum()
VentaT4 = tienda4['Precio'].sum()
```

```python
print(f"Tienda 1 vendió: {VentaT1:,}")
print(f"Tienda 2 vendió: {VentaT2:,}")
print(f"Tienda 3 vendió: {VentaT3:,}")
print(f"Tienda 4 vendió: {VentaT4:,}")
```

# 2. Ventas por categoría

```python
Cat1 = tienda.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
Cat2 = tienda2.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
Cat3 = tienda3.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
Cat4 = tienda4.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
```

```python
print(f"Ventas de Tienda 1 por {Cat1}")
print(f"Ventas de Tienda 2 por {Cat2}")
print(f"Ventas de Tienda 3 por {Cat3}")
print(f"Ventas de Tienda 4 por {Cat4}")
```

# 3. Calificación promedio de la tienda


```python
CalT1 = tienda['Calificación'].mean()
CalT2 = tienda2['Calificación'].mean()
CalT3 = tienda3['Calificación'].mean()
CalT4 = tienda4['Calificación'].mean()
```

```python
print(f"Promedio de calificación Tienda 1: {CalT1:.2f}")
print(f"Promedio de calificación Tienda 2: {CalT2:.2f}")
print(f"Promedio de calificación Tienda 3: {CalT3:.2f}")
print(f"Promedio de calificación Tienda 4: {CalT4:.2f}")
```

# 4. Productos más y menos vendidos

```python
ProdT1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ProdT2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ProdT3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ProdT4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)
```

```python
print(f"Ventas de Tienda 1 por {ProdT1}")
print(f"Ventas de Tienda 2 por {ProdT2}")
print(f"Ventas de Tienda 3 por {ProdT3}")
print(f"Ventas de Tienda 4 por {ProdT4}")
```

# 5. Envío promedio por tienda

```python
tienda['Costo de envío'] = tienda['Costo de envío'].astype(int)
tienda2['Costo de envío'] = tienda2['Costo de envío'].astype(int)
tienda3['Costo de envío'] = tienda3['Costo de envío'].astype(int)
tienda4['Costo de envío'] = tienda4['Costo de envío'].astype(int)
```

```python
EnvT1 = tienda['Costo de envío'].mean()
EnvT2 = tienda2['Costo de envío'].mean()
EnvT3 = tienda3['Costo de envío'].mean()
EnvT4 = tienda4['Costo de envío'].mean()
```

```python
print(f"Costo de envío Tienda 1: {EnvT1:,.2f}")
print(f"Costo de envío Tienda 2: {EnvT2:,.2f}")
print(f"Costo de envío Tienda 3: {EnvT3:,.2f}")
print(f"Costo de envío Tienda 4: {EnvT4:,.2f}")
```

# Graficos

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
df_Cat = pd.DataFrame({
    'Tienda 1': Cat1,
    'Tienda 2': Cat2,
    'Tienda 3': Cat3,
    'Tienda 4': Cat4
})

plt.figure(figsize=(12, 6))
for tienda in df_Cat.columns:
    plt.plot(df_Cat.index, df_Cat[tienda], marker='o', label=tienda)

plt.xticks(rotation=45)
plt.title('Cantidad de ventas por categoría en cada tienda')
plt.xlabel('Categorías')
plt.ylabel('Cantidad De Ventas')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas = [VentaT1, VentaT2, VentaT3, VentaT4]

# Convertir tiendas a posiciones numéricas
x = np.arange(len(tiendas))

# Graficar stem
plt.figure(figsize=(8, 5))
markerline, stemlines, baseline = plt.stem(x, ventas)
plt.xticks(x, tiendas)
plt.title('Ventas totales por tienda (Stem)')
plt.xlabel('Tienda')
plt.ylabel('Ventas')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.title('Costo de envío en cada tienda')
plt.ylabel('Precio')
plt.bar('Tienda 1', EnvT1)
plt.bar('Tienda 2', EnvT2)
plt.bar('Tienda 3', EnvT3)
plt.bar('Tienda 4', EnvT4)
plt.tight_layout()
plt.show()
```
